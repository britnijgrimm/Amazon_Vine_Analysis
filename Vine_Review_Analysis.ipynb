```python
#Import Dependencies
import pandas as pd
#import numpy as np
```

```python
#File to load
vine_data = "Resources/vine_table.csv"

#Read in vine_table to a Data Frame
vine_data_df = pd.read_csv(vine_data)

#Preview Data Frame
vine_data_df.head()
```

```python
#Filter data for total_votes count equal or greater to 20 
#and save to a new Data Frame
vote_data_df = vine_data_df.query('total_votes>=20')

#Preview Data Frame
vote_data_df
```

```python
#Filter new Data Frame where helpful_votes divided by total_votes is equal or greater to 50% and save to a new Data Frame
#Set vote_data_df to a new Data Frame for helpful votes
helpful_vote_df = vote_data_df

#set masks to protect data while performing calcutions
mask_helpful_votes_col = vote_data_df['helpful_votes']
mask_total_votes_col = vote_data_df['total_votes']
#mask_percent_helpful_col = helpful_vote_df['percents']

#Add new column to hold caluclated values and conduct caluation, divising helpful_votes by total_votes to obtain percentage
helpful_vote_df['percents'] = mask_helpful_votes_col / mask_total_votes_col 


#Filter for 50% or greater
helpful_vote_df = helpful_vote_df.query('percents>=.50')

#Preview Data Frame
helpful_vote_df
```

```python
#Filter new Data Frame that filters for only vine == 'Y' and save to a new Data Frame
#Set vine_data_df to a new Data Frame
yes_vine_df = vine_data_df

#Filter for only "Y" values in vine column
yes_vine_df = yes_vine_df[yes_vine_df['vine'] == 'Y']

#Preview Data Frame
yes_vine_df
```

```python
#Filter new Data Frame that filters for only vine == 'N' and save to a new Data Frame
#Set vine_data_df to a new Data Frame
no_vine_df = vine_data_df

#Filter for only "Y" values in vine column
no_vine_df = no_vine_df[no_vine_df['vine'] == 'N']

#Preview Data Frame
no_vine_df
```

```python
#Determine the total number of reviews, the number of 5-star reviews and 
#the percentage of 5-star reviews for both paid and unplaid types of reviews

#Total number of reviews
rows_df = vine_data_df['review_id']
total_rows = len(rows_df)


total_rows
```

```python
#Total number of reviews for vines
yes_rows_df = yes_vine_df['review_id']
yes_total_rows = len(yes_rows_df)

#Print count
yes_total_rows
```

```python
#Total number of reviews for non-vines
no_rows_df = no_vine_df['review_id']
no_total_rows = len(no_rows_df)

#Print count
no_total_rows
```

```python
#Total number of 5-star reviews
five_star_df = vine_data_df.query('star_rating>=5.0')
five_star_total = len(five_star_df)

#Print count
five_star_total
```

```python
#Total number of 5-star reviews for vines
five_star_yes_df = yes_vine_df.query('star_rating>=5.0')
five_star_yes = len(five_star_yes_df)

#Print count
five_star_yes
```

```python
#Total number of 5-star reviews for non-vines
five_star_no_df = no_vine_df.query('star_rating>=5.0')
five_star_no = len(five_star_no_df)

#Print count
five_star_no
```

```python
#Percentage of 5-star reviews for vines
print(five_star_yes/five_star_total)
```

```python
#Percentage of 5-star reviews for non-vines
print(five_star_no/five_star_total)
```
